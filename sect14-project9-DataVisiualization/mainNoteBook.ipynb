```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
sns.set(color_codes=True)
%matplotlib inline
```

```python
years = [1, 1000, 1500, 1600, 1700, 1750, 1800, 1850, 1900]
years = years + list(np.arange(1950, 2020, 5))

pop = [200, 400, 458, 580, 682, 791, 1000, 1261, 1650, 2525, 2758, 3018, 3322, 3682, 4061, 4440, 4853, 5310, 5735,
       6127, 6520, 6930, 7349]

plt.plot(years, pop)
```

```python
import pandas as pd
wp_over_time = pd.DataFrame(pop, index=years)
wp_over_time.plot()
```

```python
years = years[9:]
pop = pop[9:]
print(years, "\n\n", pop)
```

```python
pop_upd = []
for i in [str(x)[:2] for x in pop]:
    pop_upd.append(float(i[0] + '.' + i[1]))
    
pop = pop_upd
pop
```

```python
plt.plot(years, pop)
```

```python
pd.DataFrame(pop, index=years).plot()
```

```python
plt.plot(years, pop, color=(255/255, 100/255, 100/255))
plt.ylabel("Population in billions")
plt.xlabel("Population growth by year")
plt.title("Population Growth")
```

```python
wp_over_time = pd.DataFrame(pop, index=years)
#version with table in the graphic
wp_over_time.plot(color=(255/255, 100/255, 100/255), title="Population growth in billions", table=True)
plt.ylabel("Population in billions")
plt.xlabel("Population growth by year")
```

```python
import numpy as np
col_count = 3
index = np.arange(col_count)

korea_scores = (554, 536, 538)
canada_scores = (518, 523, 525)
china_scores = (613, 570, 580)
france_scores = (495, 505, 499)

k1 = plt.bar(index, korea_scores, .5, label="korea", color=(.3, .3, .7))
plt.grid(True)
plt.title("Korean scores")
```

```python
#using pandas frames which utilizes pyplot under the hood - with legend
cty_scores = pd.DataFrame({"Korea": korea_scores, "Canada": canada_scores, "China": china_scores, "France": france_scores},
                          index=['Mathematics', "Reading", "Science"])
cty_scores.plot(kind="bar", legend=True)
#plt.legend(loc="best")
plt.title("Country test scores")
#set the index labels to straight line
plt.xticks(rotation=0)
plt.show()
```

```python
#using pandas frames which utilizes pyplot under the hood - without legend
cty_scores = pd.DataFrame({"Korea": korea_scores, "Canada": canada_scores, "China": china_scores, "France": france_scores},
                          index=['Mathematics', "Reading", "Science"])
cty_scores.plot(kind="bar", legend=False)
#plt.legend(loc="best")
#use pyplot to label the graph
plt.title("Test Scores by Country")
plt.ylabel("Mean Score in PISA 2012")
plt.xlabel("Subjects")
#set the index labels to straight line
plt.xticks(rotation=0)
plt.show()
```
