```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
sns.set(color_codes=True)
%matplotlib inline
```

```python
years = [1, 1000, 1500, 1600, 1700, 1750, 1800, 1850, 1900]
years = years + list(np.arange(1950, 2020, 5))

pop = [200, 400, 458, 580, 682, 791, 1000, 1261, 1650, 2525, 2758, 3018, 3322, 3682, 4061, 4440, 4853, 5310, 5735,
       6127, 6520, 6930, 7349]

plt.plot(years, pop)
```

```python
import pandas as pd
wp_over_time = pd.DataFrame(pop, index=years)
wp_over_time.plot()
```

```python
years = years[9:]
pop = pop[9:]
print(years, "\n\n", pop)
```

```python
pop_upd = []
for i in [str(x)[:2] for x in pop]:
    pop_upd.append(float(i[0] + '.' + i[1]))
    
pop = pop_upd
pop
```

```python
plt.plot(years, pop)
```

```python
pd.DataFrame(pop, index=years).plot()
```

```python
plt.plot(years, pop, color=(255/255, 100/255, 100/255))
plt.ylabel("Population in billions")
plt.xlabel("Population growth by year")
plt.title("Population Growth")
```

```python
wp_over_time = pd.DataFrame(pop, index=years)
#version with table in the graphic
wp_over_time.plot(color=(255/255, 100/255, 100/255), title="Population growth in billions", table=True)
plt.ylabel("Population in billions")
plt.xlabel("Population growth by year")
```

```python
deaths = np.arange(1,4,.2)
```

```python
len(deaths)
deaths = deaths[:len(deaths)-1]
```

```python
plt.plot(years, pop, color=(255/255, 100/255, 100/255))
plt.plot(years, deaths, color=(.6, .6, 1))
plt.ylabel("Population in billions")
plt.xlabel("Population growth by year")
plt.title("Population Growth")
```

```python
#

wp_over_time = pd.DataFrame({'pop':list(pop), 'deaths': list(deaths)}, index=years)
wp_over_time.plot(title="Population growth in billions")
plt.ylabel("Population in billions")
plt.xlabel("Population growth by year")
```

```python
plt.plot(years, pop, '--', color=(255/255, 100/255, 100/255))
plt.plot(years, deaths, color=(.6, .6, 1))
plt.ylabel("Population in billions")
plt.xlabel("Population growth by year")
plt.title("Population Growth")
```

```python
lines = plt.plot(years, pop, years, deaths)
plt.grid(True)      
```

```python
lines = plt.plot(years, pop, years, deaths)
plt.grid(True)
plt.setp(lines, color=(1, .4, .4), marker='o')
plt.show()
```

```python
labels = 'Python', 'C++', 'Ruby', 'Java', 'PHP', 'Perl'
sizes = [33, 52, 12, 17, 62, 48]
seperated = [.1, .1, 0, 0, 0, 0] #to use float a section
```

```python
df_pie = pd.DataFrame({'labels': labels, 'size': sizes})
df_pie.plot(kind="pie", y='size', labels=labels, legend=False, autopct='%1.1f%%')
plt.axis('equal')
plt.show()
```

```python
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.axis('equal')
plt.show()
```

```python
df_pie = pd.DataFrame({'labels': labels, 'size': sizes})
df_pie.plot(kind="pie", y='size', labels=labels, legend=False, autopct='%1.1f%%', explode=seperated)
plt.axis('equal')
plt.show()
```

```python
plt.pie(sizes, labels=labels, autopct='%1.1f%%', explode=seperated)
plt.axis('equal')
plt.show()
```
